```python
import pandas as pd
import nlplot
import plotly.express as px
```

```python
# read source file
reviews_df = pd.read_csv("Resources/all_reviews_data.csv", encoding='ISO-8859-1')


# count of movie plot summaries
# len(corpus)
reviews_df
```

```python
npt = nlplot.NLPlot(reviews_df, target_col='Review_Text')
```

```python
stopwords = npt.get_stopword(top_n=30, min_freq=0)
```

```python
npt.bar_ngram(title='uni-gram', ngram=1, top_n=50, stopwords=stopwords)
npt.bar_ngram(title='bi-gram', ngram=2, top_n=50, stopwords=stopwords)
```

```python
npt.treemap(title='Tree of Most Common Words', ngram=1, top_n=30, stopwords=stopwords)
```

```python
wordhist = npt.word_distribution(title='words distribution')
wordhist.write_html("Resources/wordhistogram.html")
wordhist
```

```python
nptcloud = npt.wordcloud(stopwords=stopwords, colormap='tab20_r')
nptcloud.write_html("Resources/nptcloud.html")
```

```python
npt.build_graph(stopwords=stopwords, min_edge_frequency=10)
# The number of nodes and edges to which this output is plotted.
# If this number is too large, plotting will take a long time, so adjust the [min_edge_frequency] well.
edge_size = 166
codep = npt.co_network( title='Co-occurrence network', node_size='adjacency_frequency')
codep
```

```python
npt.sunburst(title='sunburst chart', colorscale=True)
```
